```python
import pathlib
import sys

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
parquet_files_path = pathlib.Path("../../2.cellprofiling/results/C4-2").resolve(
    strict=True
)
parquet_files = sorted(parquet_files_path.glob("*.parquet"))
df = pd.read_parquet(parquet_files[0])
df1 = pd.read_parquet(parquet_files[1])
# if object_id in column name then rename it to object_id
for col in df.columns:
    if "object_id" in col:
        df.rename(columns={col: "object_id"}, inplace=True)
        break

for col in df1.columns:
    if "object_id" in col:
        df1.rename(columns={col: "object_id"}, inplace=True)
        break
```

```python
feature_types_dict = {
    "AreaSize_Shape": [],
    "Colocalization": [],
    "Intensity": [],
    "Granularity": [],
    "Neighbor": [],
    "Texture": [],
}
```

```python
for file in parquet_files:
    for feature_type in feature_types_dict.keys():
        if feature_type in file.stem:
            feature_types_dict[feature_type].append(file)
feature_types_dict.keys()
```

```python
area_size_shape_files = [
    pd.read_parquet(file) for file in feature_types_dict["AreaSize_Shape"]
]
# merge all dataframes in the list on the object_id column
final_df = pd.DataFrame()
for file in feature_types_dict["AreaSize_Shape"]:
    df = pd.read_parquet(file)
    for col in df.columns:
        if "object_id" in col:
            df.rename(columns={col: "object_id"}, inplace=True)
            break
    if not final_df.empty:
        final_df = pd.merge(final_df, df, on=["object_id", "image_set"], how="outer")
    else:
        final_df = df
final_df
```

```python
colocalization_files = [
    pd.read_parquet(file) for file in feature_types_dict["Colocalization"]
]
colocalization_df = pd.concat(colocalization_files, ignore_index=True)
intensity_files = [pd.read_parquet(file) for file in feature_types_dict["Intensity"]]
intensity_df = pd.concat(intensity_files, ignore_index=True)
# granularity_files = [pd.read_parquet(file) for file in feature_types_dict["Granularity"]]
# granularity_df = pd.concat(granularity_files, ignore_index=True)
# neighbor_files = [pd.read_parquet(file) for file in feature_types_dict["Neighbor"]]
# neighbor_df = pd.concat(neighbor_files, ignore_index=True)
# texture_files = [pd.read_parquet(file) for file in feature_types_dict["Texture"]]
# texture_df = pd.concat(texture_files, ignore_index=True)
```
